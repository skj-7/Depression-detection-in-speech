```python
import speech_recognition as sr 
import os 
from pydub import AudioSegment
from pydub.silence import split_on_silence


r = sr.Recognizer()

def get_audio_text(path):
    
    folder_name = "./audio-chunks"
    sound = AudioSegment.from_wav(path)  
    chunks = split_on_silence(sound,min_silence_len = 500,silence_thresh = sound.dBFS-14,keep_silence=500)
    
    if not os.path.isdir(folder_name):
        os.mkdir(folder_name)
                              
    transcription = ""
    
    for i, audio_chunk in enumerate(chunks, start=1):

        chunk_filename = os.path.join(folder_name, f"chunk{i}.wav")
        audio_chunk.export(chunk_filename, format="wav")

        with sr.AudioFile(chunk_filename) as source:
            audio_listened = r.record(source)

            try:
                text = r.recognize_google(audio_listened)
            except sr.UnknownValueError as e:
                pass
            except sr.RequestError as e:
                pass
            else:
                text = f"{text.capitalize()}. "
                transcription += text

    print(transcription)            
    return transcription
```

**CSV Formation**

```python
import csv

file = open('audio_text.csv','w')
writer = csv.writer(file)

header = ['Person','Audio_text']
writer.writerow(header)

for audio_file in os.listdir("../input/mp-audio-dataset-100-entries/data1"):
    file_path = "../input/mp-audio-dataset-100-entries/data1/"+audio_file+'/'+audio_file[:3]+'_AUDIO.wav'
    writer.writerow([audio_file[:3],get_audio_text(file_path)])
file.close()
```
