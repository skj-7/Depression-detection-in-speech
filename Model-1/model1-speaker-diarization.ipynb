```python
from pyannote.audio import Pipeline
from pydub import AudioSegment
import os

pipeline = Pipeline.from_pretrained('pyannote/speaker-diarization',use_auth_token='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')

os.mkdir('./audio_data_seg')

for audio_file in os.listdir('../input/mp-audio-dataset-100-entries/data1'):
    file_path = "../input/mp-audio-dataset-100-entries/data1/"+audio_file+'/'+audio_file[:3]+'_AUDIO.wav'
    diarization = pipeline(file_path, num_speakers=2)
    
    sound = AudioSegment.from_wav(file_path)
    
    for turn, _, speaker in diarization.itertracks(yield_label=True):
        extract = sound[turn.start*1000 : turn.end*1000]
        seg_name = audio_file[:3]+ f"start={turn.start:.1f}s stop={turn.end:.1f}s speaker_{speaker}.wav"
        extract.export("./audio_data_seg/"+seg_name, format="wav")
```
